```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df_model = df.drop('Job Title', axis=1)
df_model
```

```python
df_model.hist(figsize=(15,8))
```

```python
log_transform_features = ['Age', 'Gender', 'Education Level', 'Years of Experience']

for feature in log_transform_features:
    df_model[feature] = np.log(df_model[feature] + 1)
    
df_model.hist(figsize=(15,8))
```

```python
df_model.dropna(inplace=True)
df_model
```

```python
X = df_model.drop(['Salary'], axis=1)
y = df_model['Salary']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
train_data = X_train.join(y_train)
test_data = X_test.join(y_test)

X_train, y_train = train_data.drop(['Salary'], axis=1), train_data['Salary']
x_test, y_test = test_data.drop(['Salary'], axis=1), test_data['Salary']

reg = LinearRegression()
reg.fit(X_train, y_train)

reg.score(x_test, y_test)
```
